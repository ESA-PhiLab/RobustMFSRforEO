```python
import sys
sys.path.append('..')
import pandas as pd
from shapely.geometry import Polygon
from dataset_tools.utils import getCoords
from dataset_tools.Sentinel2Loader import Sentinel2Loader
import os
import shutil
from datetime import datetime
from datetime import timedelta
images = ["ID_0_2019-10-11.tif","ID_2_2019-10-14.tif","SD_1_2018-06-08.tif","SD_3_2018-10-19.tif","UC_1_2018-05-25.tif",
          "ID_1_2019-09-25.tif","SD_0_2018-06-05.tif","SD_2_2018-06-27.tif","UC_0_2018-05-25.tif"]
dates = []
dates_beg = []
dates_end = []
for i in images:
    dates += [i[5:-4]]
    dateObj = datetime.strptime(i[5:-4], '%Y-%m-%d')
    dates_beg += [(dateObj-timedelta(15)).strftime("%Y-%m-%d")]
    dates_end += [(dateObj+timedelta(15)).strftime("%Y-%m-%d")]

root_dir = "/data/wotus/Planet/"
sl = Sentinel2Loader('/home/muhammed/sentinelcache', 
                    'mtrazzak', 'zivhu4-sortud-kuJxiv',
                    apiUrl='https://scihub.copernicus.eu/apihub/', showProgressbars=False, cloudCoverage=(0,100))
if not os.path.exists("/data/wotus/sentinel/"):
    os.makedirs("/data/wotus/sentinel/")
for idx, im in enumerate(images):
    area = Polygon(getCoords(root_dir+im))
    geoTiffs = sl.getRegionHistory(area, 'TCI','10m', dates_beg[idx], dates_end[idx], daysStep=1, aoi=im[:-4])
    for geoTiff in geoTiffs:
        name = os.path.basename(geoTiff)
        shutil.move(geoTiff, "/data/wotus/sentinel/"+name)
    print(im)
    print(str(idx+1)+"/"+str(len(images)))
```
